```python
import requests
url='https://opendata.cwb.gov.tw/api/v1/rest/datastore/F-C0032-001?Authorization=CWB-23A106BA-F626-4A91-971F-AF743D85A4C5'
data = requests.get(url)  
data_json = data.json() 
for i in data_json['records']['location']:
    city=i['locationName']
    wx8 = i['weatherElement'][0]['time'][0]['parameter']['parameterName'] 
    min8 = i['weatherElement'][2]['time'][0]['parameter']['parameterName'] 
    max8 = i['weatherElement'][4]['time'][0]['parameter']['parameterName'] 
    # ci = i['weatherElement'][3]['time'][0]['parameter']['parameterName']
    pop = i['weatherElement'][1]['time'][0]['parameter']['parameterName']  
    print(city,wx8,"最低溫",min8,"℃,最高溫",max8,"℃,降雨機率",pop,"%")
```
